# Data Visualization

## Opening the csv files

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Tagalog to Ilokano

```python
# Read the data
df_bleu_sd_tl_il = pd.read_csv('../src/scores/Tagalog to Ilokano/Standard Translator/test/test_bleu_tl-il.csv')
df_bleu_hb_tl_il = pd.read_csv('../src/scores/Tagalog to Ilokano/Hybrid Translator/test/test_bleu_tl-il.csv')
```

```python
df_bleu_hb_tl_il.head()
```

# Creating the vizualization

```python
# Comparing the histograms of the two models
fig, ax = plt.subplots(1, 2, figsize=(15, 5))
sns.histplot(df_bleu_sd_tl_il['BLEU_SCORE'], ax=ax[0], color='blue')
sns.histplot(df_bleu_hb_tl_il['BLEU_SCORE'], ax=ax[1], color='red')
ax[0].set_title('Standard Translator')
ax[1].set_title('Hybrid Translator')
fig.suptitle('BLEU Score Distribution (Tagalog to Ilokano)')
plt.show()
```
