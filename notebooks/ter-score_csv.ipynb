# TER Scoring TL-IL (ALL)


```python
import pandas as pd
import sacrebleu

# Read the tagalog POS dataset
dict_tl_il_result = pd.read_json('../src/json data/Tagalog to Ilokano/Hybrid Translator/dict_op_ex.json')

source_text_list = dict_tl_il_result['Source Text'].tolist()
system_op_list = dict_tl_il_result['System Output'].tolist()
target_op_list = dict_tl_il_result['Target Output'].tolist()
```

```python
with open("../src/scores/csv/ter_scores.csv","w", encoding="utf-8") as df:
    temp_index = 0
    total_ter_score = 0
    score_list = []

    target = [['']]
    system = ['']
    data = []
    df = pd.DataFrame(data, columns=["INPUT", "REFERENCE", "MACHINE_TRANSLATION", "TER_Score"])

    for target_op in target_op_list:
        system_op = system_op_list[temp_index]
        source_text = source_text_list[temp_index]
        system[0] = system_op
        target[0] = [target_op]
        
        score = str(sacrebleu.corpus_ter(system, target))

        for i in score.split():
            try:
                #trying to convert i to float
                result = float(i)
                #break the loop if i is the first string that's successfully converted
                break
            except:
                continue
            
        """
            Writing the result to a csv file
        """

       
        
        df = df.append({"INPUT": source_text, "REFERENCE": target_op, "MACHINE_TRANSLATION": system_op, "TER_Score": str(result)}, ignore_index=True)
        df.to_csv('../src/scores/csv/ter_scores.csv', index=False)
        
        score_list.append(result)
        total_ter_score += result
        temp_index += 1
        
average_ter_score = total_ter_score / len(target_op_list)
df=  df.append({"Average TER_Score": str(average_ter_score)}, ignore_index=True)
df.to_csv('../src/scores/csv/ter_scores.csv', index=False)

```
