```python
import pandas as pd


verb_dict = pd.read_json('src/json data/verb_dict.json')

verb_dict.head()
```

```python
import csv

field_names= ['Salitang-ugat', 'Pangnagdaan', 'Pangkasalukuyan', 'Panghinaharap', 'Pawatas', 'Katatapos']

with open('src/csv data/Verb.csv', 'w') as csvfile:
    writer = csv.DictWriter(csvfile, fieldnames=field_names)
    writer.writeheader()
    writer.writerows(verb_dict.to_dict('records'))
    
```

```python
def trim(dataset):
    trim = lambda x: x.strip() if type(x) is str else x
    return dataset.applymap(trim)

dataset = trim(pd.read_csv('src/csv data/Verb.csv'))
dataset.head()
```
